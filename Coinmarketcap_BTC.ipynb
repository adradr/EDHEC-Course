```python
import quandl
import pandas as pd
```

```python
btc = quandl.get("BITFINEX/BTCUSD")
eth = quandl.get("BITFINEX/ETHUSD")
```

```python
l = ["Last"]
```

```python
data = pd.DataFrame()
```

```python
data = btc[l]
```

```python
data = data.rename(columns={"Last": "BTC"})
```

```python
data["ETH"] = eth[l]
```

```python
data_change = data.pct_change()
```

```python
data_change
```

```python
data_change.plot()
```

```python
data.plot()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some mock data
t = data.index
data1 = data["BTC"]
data2 = data["ETH"]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time')
ax1.set_ylabel('BTC', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('ETH', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
